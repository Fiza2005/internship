```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=df.copy()
```

```python
encoded_data
```

```python
label_encoder=LabelEncoder()
```

```python
encoded_data['Gender_Label']=label_encoder.fit_transform(encoded_data['Gender'])
```

```python
one_hot_encoded=pd.get_dummies(encoded_data['Education'],prefix='Education')
```

```python
encoded_data = pd.concat([encoded_data, one_hot_encoded],axis=1)
```

```python
Performance_Rating_mapping= {1: 'First', 2: 'Second', 3: 'Third',4:'Fourth',5:'Fifth'}
```

```python
encoded_data['Performance_Rating_Category'] = encoded_data['Performance_Rating'].map(Performance_Rating_mapping)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
path=pd.read_csv('/content/preprocessing_practice_dataset.csv')
```

```python
numerical_columns=['Age','Experience_Years','Monthly_Salary']
```

```python

path.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
path['Age_MinMax']=minmax_scaler.fit_transform(path[['Age']])
path['Experience_Years_MinMax']=minmax_scaler.fit_transform(path[['Experience_Years']])
```

```python
path['Age_Standard']=standard_scaler.fit_transform(path[['Age']])
path['Experience_Years_Standard']=standard_scaler.fit_transform(path[['Experience_Years']])
```
